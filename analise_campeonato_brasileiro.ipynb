
#  An谩lise de Dados do Campeonato Brasileiro

Neste notebook, exploramos e analisamos os dados do Campeonato Brasileiro. A an谩lise abrange v谩rias m茅tricas, como desempenho dos times, vit贸rias por mandante e visitante, entre outras estat铆sticas interessantes. Utilizamos a biblioteca **Pandas** para manipular os dados.

## Passo 1: Carregar e Estruturar os Dados

Primeiramente, carregaremos os dados e faremos algumas prepara莽玫es, como a cria莽茫o de colunas que indicam o saldo de gols e o resultado dos mandantes.


```python

import pandas as pd

# Carregar o arquivo CSV no Pandas
df = pd.read_csv('campeonato-brasileiro-full.csv')

# Converter a coluna 'data' para o formato datetime
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y', errors='coerce')

# Criar a coluna 'saldo_de_gols' (diferen莽a entre mandante_Placar e visitante_Placar)
df['saldo_de_gols'] = df['mandante_Placar'] - df['visitante_Placar']

# Criar a coluna 'resultado_mandante' (Vit贸ria, Empate ou Derrota)
df['resultado_mandante'] = df.apply(
    lambda row: 'Vit贸ria' if row['saldo_de_gols'] > 0 else 'Derrota' if row['saldo_de_gols'] < 0 else 'Empate',
    axis=1
)

# Exibir as primeiras linhas do DataFrame
df.head()
```


## Passo 2: Desempenho Geral dos Times

Agora vamos analisar o desempenho geral dos times, come莽ando pela contagem de jogos como mandante e visitante, e a contagem de vit贸rias para mandante e visitante.


```python

# Contagem de jogos como mandante e visitante
jogos_mandante = df['mandante'].value_counts().sort_values(ascending=False)
jogos_visitante = df['visitante'].value_counts().sort_values(ascending=False)

print("Jogos como mandante: ", jogos_mandante.head())
print(" Jogos como visitante: ", jogos_visitante.head())
```

```python

# Contagem de vit贸rias dos mandantes e visitantes
vitorias_mandante = df[df['resultado_mandante'] == 'Vit贸ria']['mandante'].value_counts()
vitorias_visitante = df[df['vencedor'] == df['visitante']]['visitante'].value_counts()

print("Vit贸rias como mandante:", vitorias_mandante.head())
print("Vit贸rias como visitante:", vitorias_visitante.head())
```


## Passo 3: An谩lise dos Resultados

Agora vamos analisar a distribui莽茫o dos resultados dos jogos (vit贸ria, empate e derrota), assim como os placares mais comuns registrados nas partidas.


```python

# Distribui莽茫o de resultados (mandantes)
resultados = df['resultado_mandante'].value_counts()
print("Distribui莽茫o de resultados (Mandantes): ", resultados)

# Placar mais comum
placares_comuns = df.groupby(['mandante_Placar', 'visitante_Placar']).size().sort_values(ascending=False)
print(" Placar mais comum: ", placares_comuns.head())
```


## Passo 4: Desempenho Temporal

Vamos agora observar o desempenho dos times ao longo do tempo, por rodada e ano. Isso pode nos ajudar a identificar tend锚ncias ao longo da temporada ou entre diferentes temporadas.


```python

# Desempenho por rodada (vit贸rias dos mandantes)
rodadas_vitorias = df[df['resultado_mandante'] == 'Vit贸ria'].groupby('rodata').size()
print("Vit贸rias por rodada (mandante): ", rodadas_vitorias.head())

# Desempenho por ano (vit贸rias dos mandantes)
df['ano'] = df['data'].dt.year
vitorias_ano = df.groupby('ano')['resultado_mandante'].apply(lambda x: (x == 'Vit贸ria').sum())
print("Vit贸rias por ano: ", vitorias_ano)
```


## Passo 5: An谩lise Geogr谩fica

Nesta an谩lise, vamos ver como os times de diferentes estados se comportam, analisando vit贸rias de mandantes e visitantes por estado.


```python

# Vit贸rias por estado (mandantes)
vitorias_estado = df[df['resultado_mandante'] == 'Vit贸ria'].groupby('mandante_Estado').size().sort_values(ascending=False)
print("Vit贸rias por estado (mandantes): ", vitorias_estado)

# Vit贸rias fora do estado (visitantes)
vitorias_fora_estado = df[df['vencedor'] == df['visitante']].groupby('visitante_Estado').size().sort_values(ascending=False)
print(" Vit贸rias fora do estado (visitantes): ", vitorias_fora_estado)
```


## Passo 6: Arena e Local dos Jogos

Agora vamos analisar a influ锚ncia das arenas nos resultados, observando as vit贸rias e os gols registrados em cada arena.


```python

# Arenas com mais vit贸rias dos mandantes
vitorias_arena = df[df['resultado_mandante'] == 'Vit贸ria'].groupby('arena').size().sort_values(ascending=False)
print("Arenas com mais vit贸rias dos mandantes: ", vitorias_arena.head())

# Arenas com mais gols
gols_arena = df.groupby('arena')['mandante_Placar', 'visitante_Placar'].sum()
print(" Arenas com mais gols: ", gols_arena.head())
```


## Passo 7: An谩lise de Gols

Nesta se莽茫o, vamos explorar as m茅dias de gols dos mandantes e visitantes, al茅m de identificar jogos com goleadas (diferen莽a de gols maior ou igual a 3).


```python

# M茅dia de gols dos mandantes e visitantes
media_gols_mandante = df['mandante_Placar'].mean()
media_gols_visitante = df['visitante_Placar'].mean()
print(f"M茅dia de gols dos mandantes: {media_gols_mandante}")
print(f"M茅dia de gols dos visitantes: {media_gols_visitante}")


goleadas = df[(df['mandante_Placar'] - df['visitante_Placar']).abs() >= 3]

print(f"N煤mero de goleadas (diferen莽a >= 3 gols): {goleadas.shape[0]}")
```


## Passo 8: Outras Estat铆sticas Interessantes

Finalmente, vamos explorar algumas estat铆sticas adicionais, como o jogo com o maior n煤mero de gols e a quantidade de empates sem gols (0x0).


```python

# Jogo com o maior n煤mero de gols
jogo_com_mais_gols = df[df['mandante_Placar'] + df['visitante_Placar'] == (df['mandante_Placar'] + df['visitante_Placar']).max()]
print("Jogo com maior n煤mero de gols: ", jogo_com_mais_gols[['mandante', 'visitante', 'mandante_Placar', 'visitante_Placar']])

# N煤mero de empates por 0x0
empates_sem_gols = df[(df['mandante_Placar'] == 0) & (df['visitante_Placar'] == 0)].shape[0]
print(f"N煤mero de empates por 0x0: {empates_sem_gols}")

# N煤mero de jogos com viradas
viradas = df[(df['saldo_de_gols'] < 0) & (df['vencedor'] == df['visitante'])].shape[0]
print(f"Jogos com viradas (vit贸ria do visitante): {viradas}")
```
